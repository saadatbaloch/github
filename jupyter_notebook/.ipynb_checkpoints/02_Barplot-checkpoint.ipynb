### Adding Estimator
- You have to put one numerical veriable on y axis because catagorical veriable cannot be estimated
- Estimation k lye aik library use hoti hy jisko Numpy kehty hain

```python
# import libraries
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

# load dataset
titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot
sns.barplot(x="class", y="fare", hue="sex", data=titanic, estimator=median)
plt.title("kon kon doobny sy bacha")
plt.show()
```

### Calculating mean


```python
# import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset
titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot
sns.barplot(x="class", y="fare", hue="sex", data=titanic, estimator=mean, saturation=0.5)
plt.title("kon kon doobny sy bacha")
plt.show()
```

```python
# import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset
titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot
sns.barplot(x="fare", y="class", hue="sex", data=titanic, estimator=mean, saturation=1)
plt.title("kon kon doobny sy bacha")
plt.show()
```

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt

# read a titanic.csv file
# from seaborn library 
titanic = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", data=titanic,
           linewidth=2.5, facecolor=(0.1,0.3,0.3,0.5),
           errcolor=".5", edgecolor="0")
```
